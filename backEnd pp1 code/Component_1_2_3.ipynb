```python
# data cleaning - Remove punctuation marks

punctuation = '''!()@#&:;ЁЯдЬЁЯдЫЁЯОВЁЯШОЁЯТЪ...тЭд/%?\{}$,-ЁЯШВЁЯЩП*ЁЯСЗЁЯФеЁЯШвЁЯС╣ЁЯТШЁЯШКЁЯСНЁЯСОЁЯСКЁЯШгЁЯдХЁЯШИЁЯШДтШаЁЯШн1234567890ЁЯШАЁЯТЧЁЯе░ЁЯШНЁЯТпЁЯдйтЩея╕ПЁЯдШЁЯСИтЬМя╕ПЁЯЩВЁЯТктЬКЁЯП┐ЁЯШШЁЯе│ЁЯдЯЁЯШлЁЯШКЁЯТЦя╕ПЁЯМ╣ЁЯСМЁЯе│ЁЯдЧЁЯдоЁЯдвЁЯОмЁЯе│ЁЯдпЁЯШЗЁЯдгЁЯП╜ЁЯШГЁЯТеЁЯдФ'''
def remove_punct(cmt):
    cmt_nonpunct = "".join([ char for char in cmt if char not in punctuation])  #--- for char in cmt: ----
                                                                                #---   if( char not in punctuation ): ----                               
    return cmt_nonpunct

xl_data['Cleaned_Comments'] = xl_data['Original_Comments'].apply(lambda x: remove_punct(x))
xl_data
```

```python
#Language detection & translate to common-language(english)

translated_words = []
src_language = []
dest_language = []

for element in xl_data['Cleaned_Comments']:
    translated_words.append(translator.translate(element).text)
    src_language.append(LANGUAGES[ translator.translate(element).src] )
    dest_language.append(LANGUAGES[ translator.translate(element).dest] )
    
# print(translated_words)
# print(src_language)
# print(dest_language)
    
xl_data['Source_Language'] = src_language
xl_data['destination_Language'] = dest_language
xl_data['English_Comments'] = translated_words

xl_data
```

```python
# Tokenization

def tokenize(cmt):
    tokens = re.split('\W+', cmt)
    return tokens

xl_data['tokens'] = xl_data['English_Comments'].apply(lambda x: tokenize(x))
xl_data
```

```python
# remove unwanted words

remove_words = []

remove_words.append('httpswwwyoutubecomwatchv')
remove_words.append('yjlRWSiampt')
remove_words.append('href')
remove_words.append('br')
remove_words.append('Bar')
remove_words.append('Kahatiti')
remove_words.append('ms')
remove_words.append('a')
remove_words.append('the')
remove_words.append('tha')

def remove_stop_words(cmt_tokenize):
    clean_cmt = [word for word in cmt_tokenize if word not in remove_words]
    return clean_cmt

xl_data['cleaned_tokens'] = xl_data['tokens'].apply(lambda x: remove_stop_words(x))
xl_data
```

```python
# merge tokens

def merge_tokens(cmt):
    merge_token = " ".join(cmt)                                                                                                          
    return merge_token

xl_data['merge_tokens'] = xl_data['cleaned_tokens'].apply(lambda x: merge_tokens(x))
xl_data 
```

```python
# Translate to tamil

translated_tamil_words = []

for element in xl_data['merge_tokens']:
    translated_tamil_words.append(translator.translate(element, dest="ta").text)
    
xl_data['Tamil_Comments'] = translated_tamil_words
xl_data
```

```python
# Create List for positive & negative Comments

positive_list = ['nanraka', 'alakana', 'siranthathu', 'verupadda', 'virumpukiren', 'olukkamana', 'nalla', 'sirantha', 'suppar', 'maas', 'rasikan', 'geya', 'veramathi', 'nalla', 'senjudar', 'nallam', 'paarkalaam', 'senjudar', 'senjudanga', 'neruppu', 'mersal', 'next level', 'level', 'vera kaddam', 'verelevel']
negative_list = ['virayam', 'kothikalan', 'keli', 'kanravi', 'kevalam', 'asinkam', 'kevalamana', 'asinkamana', 'bore', 'boomer', 'beep', 'nallave illa', 'nalla illa', 'poor', 'mayiru', 'bad', 'failure', 'fail', 'sleep']
```

```python
# calculation for rating

positive_comments_count = 0
negative_comments_count = 0

for cmt in data_set['tanglish_token']:  
            try:  
                for word in cmt:
                    if word.lower() in positive_list:
                        while word.lower() not in positive_list:
                            continue
                        positive_comments_count += 1 
#                         print(word)
                        break

                for word in cmt:
                    if word.lower() in negative_list:
                        while word.lower() not in negative_list:
                            continue
                        negative_comments_count += 1
#                         print(word)
                        break
            except:
                pass
    
print('Positive comments count : ', positive_comments_count )
print('Negative comments count : ', negative_comments_count )
```

```python
# get rating

positive_rating = positive_comments_count/(positive_comments_count + negative_comments_count)
print( 'Positive Comment = ',  round(positive_rating*100, 2), '%')

negative_rating = negative_comments_count/(positive_comments_count + negative_comments_count)
print( 'Negative Comment = ', round(negative_rating*100, 2), '%')
```

```python
# making decision

if positive_rating > negative_rating:
    print('This movie has more "positive" review than negative review...')
else:
    print('This movie has more "negative" review than positive review...')
```

```python
# Create list for comedy, action & love

comedy_list = ['sirippu', 'siriththan', 'comedy', 'nakaisuvai']

action_list = ['sandai', 'sandaikal', 'sandaikasikal', 'thi', 'Kothikalan', 'Athiradi', 'kathaikkalam']

love_list = ['kathal', 'lovers', 'kathalarkal', 'kathalar', 'kaviyam', 'virumpukiren', 'olukkamana', 'alakana']

horror_list = ['payama', 'payanthudan', 'thikil', 'pei', 'payankaram']
```

```python
# calculation for rating

comedy_comments_count = 1
action_comments_count = 1
love_comments_count = 1
horror_comments_count = 1

for cmt in data_set['tokenize_cmt']:  
            try:  
                for word in cmt:
                    if word.lower() in comedy_list:
                        while word.lower() not in comedy_list:
                            continue
                        comedy_comments_count += 1 
                        # print(word)
                        break

                for word in cmt:
                    if word.lower() in action_list:
                        while word.lower() not in action_list:
                            continue
                        action_comments_count += 1
                        # print(word)
                        break

                for word in cmt:
                    if word.lower() in love_list:
                        while word.lower() not in love_list:
                            continue
                        love_comments_count += 1
                        # print(word)
                        break

                for word in cmt:
                    if word.lower() in horror_list:
                        while word.lower() not in horror_list:
                            continue
                        horror_comments_count += 1
                        # print(word)
                        break
            except:
                pass
    
print('Action comments count : ', action_comments_count )
print('Comedy comments count : ', comedy_comments_count )
print('Love comments count : ', love_comments_count )
print('Horror comments count : ', horror_comments_count )
```

```python
# get rating

action = action_comments_count/(action_comments_count + comedy_comments_count + love_comments_count + horror_comments_count)
print( 'Action Comment = ',  round(action*100, 2), '%')
comedy = comedy_comments_count/(action_comments_count + comedy_comments_count + love_comments_count + horror_comments_count)
print( 'Comedy Comment = ',  round(comedy*100, 2), '%')

love = love_comments_count/(action_comments_count + comedy_comments_count + love_comments_count + horror_comments_count)
print( 'Love Comment = ',  round(love*100, 2), '%')

horror = horror_comments_count/(action_comments_count + comedy_comments_count + love_comments_count + horror_comments_count)
print( 'Horror Comment = ',  round(horror*100, 2), '%')


rating = [action, comedy, love, horror]
```
